```python
import random

# Define base salaries by role
salary_by_role = {
    "Analyst": 12000,
    "Engineer": 15000,
    "Manager": 18000,
    "Intern": 5000,
    "Consultant": 10000
}

# Number of workers
num_workers = 400
roles = list(salary_by_role.keys())
genders = ["Male", "Female"]

# Generate worker list
workers = []

for i in range(1, num_workers + 1):
    role = roles[i % len(roles)]
    gender = random.choice(genders)
    salary = salary_by_role[role]
    
    worker = {
        "id": i,
        "name": f"Worker_{i}",
        "email": f"worker{i}@example.com",
        "role": role,
        "department": f"Department_{(i % 10) + 1}",
        "gender": gender,
        "salary": salary
    }
    workers.append(worker)

# Generate payment slips with employee level logic and exception handling
for worker in workers:
    try:
        name = worker['name']
        role = worker['role']
        department = worker['department']
        email = worker['email']
        gender = worker['gender']
        salary = worker['salary']
        employee_level = "N/A"

        # Apply employee level rule
# Check if salary is greater than $10,000 and less than $20,000
        if salary > 10000 and salary < 20000:
         employee_level = "A1"

        if salary > 7500 and salary < 30000 and gender == "Female":
            employee_level = "A5-F"


        print(f"--- Payment Slip ---")
        print(f"Name: {name}")
        print(f"Email: {email}")
        print(f"Gender: {gender}")
        print(f"Department: {department}")
        print(f"Role: {role}")
        print(f"Monthly Salary: ${salary}")
        print(f"Employee Level: {employee_level}")
        print("--------------------\n")

    except KeyError as e:
        print(f"Missing data for worker {worker.get('id', 'Unknown')}: {e}")
    except Exception as e:
        print(f"Unexpected error for worker {worker.get('id', 'Unknown')}: {e}")
```
